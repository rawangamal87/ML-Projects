# ***Data Preprocessing***

```python
data.drop('Id', axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler                  # rescale the data to 0 and 1
data_scaler = MinMaxScaler()
data_scaler.fit(data)
data_scaler.transform(data)
```

```python
from sklearn.preprocessing import StandardScaler      #to rescale mean to 0 and std to 1
data_scale = StandardScaler()
data_scale.fit(data)
data_scale.transform(data)
```

# ***Splitting Data***

```python
x = data.iloc[ : ,1:5]
x
```

```python
y = data.iloc[ : , 5:6]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# ***Modeling***

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.datasets import make_moons
x,y = make_moons(n_samples=500, noise=0.30, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

# ***Model Evaluation***

```python
from sklearn.metrics import confusion_matrix , classification_report
```

```python
confusion_matrix(y_test, model.predict(x_test))
```

```python
classification_report(y_test, model.predict(x_test))
```

```python
model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, model.predict(x_test))
```
