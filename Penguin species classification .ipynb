# ***Data Preprocessing***

```python
data['island'].unique()
```

```python
data['species'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['species'] = le.fit_transform(df['species'])
```

```python
df['island'] = le.fit_transform(df['island'])
```

```python
df['culmen_length_mm'] = pd.to_numeric(df['culmen_length_mm'], errors='coerce')
```

```python
df['culmen_depth_mm'] = pd.to_numeric(df['culmen_depth_mm'], errors='coerce')
```

```python
df['flipper_length_mm	'] = pd.to_numeric(df['flipper_length_mm'], errors='coerce')
```

```python
df['body_mass_g'] = pd.to_numeric(df['body_mass_g'], errors='coerce')
```

# ***Splitting Data***

```python
x = df.iloc[ : , 1: ]
x
```

```python
y = df.iloc[ : , 0:1]
y
```

```python
df.isnull().sum()
```

```python
x.fillna(df.mean(), inplace= True)
```

```python
x.isnull().sum()
```

```python
y.fillna(df.mean(), inplace= True)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.7, test_size= 0.3, random_state= 40)
```

```python
df
```

# ***Scaling Data***

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

```python
x_train_scaled
```

# ***Modeling***

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=100, criterion= 'entropy')
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)
```

```python
dt.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rc =RandomForestClassifier(n_estimators=5, min_samples_split=5, random_state= 30)
rc.fit(x_train, y_train)
```

```python
rc.score(x_train, y_train)
```

```python
rc.score(x_test, y_test)
```
