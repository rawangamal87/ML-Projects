# **label data**

```python
data.loc[data['Category'] == 'spam', 'Category',] = 0      #spam data = 0
data.loc[data['Category'] == 'ham', 'Category',] = 1       #ham data = 1
```

```python
x = data['Message']
y = data['Category']
```

```python
print(x)
```

```python
print(y)
```

# **splitting data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Feature Extraction**

```python
from sklearn.feature_extraction.text import CountVectorizer    #transform text data to feature vectors
vector = CountVectorizer()
vector.fit(x_train)
x_train_vector = vector.transform(x_train)
x_test_vector = vector.transform(x_test)
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase= True)
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)
```

```python
try:
    y_train = y_train.astype('int')
    y_test = y_test.astype('int')
except ValueError as e:
    print("Error during conversion:", e)

print("y_train after conversion:", y_train)
print("y_test after conversion:", y_test)
```

```python
print(x_train)
```

# **Training the model**

```python
model = LogisticRegression()
```

```python
model.fit(x_train_features, y_train)
```

```python
model.score(x_train_features, y_train)
```

# **Evaluating the trained model**

```python
#prediction on training data
model.predict(x_train_features)
```

```python
accuracy_on_training_data = accuracy_score(y_train, model.predict(x_train_features))
```

```python
print('accuracy on training data is', accuracy_on_training_data)
```

```python
#prediction on testing data
prediction_on_test_data = model.predict(x_test_features)
accuracy_on_test_data =  accuracy_score(y_test, prediction_on_test_data)
```
